```python
features = ['Category1','Category2', 'Category3', 'year_camp', 'year_camp_end', 'month_camp', 'month_camp_end']
```

```python
def one_hot(temp):
    x = map(lambda x: pd.get_dummies(temp[x],prefix = x),features)
    x = pd.concat(x,axis = 1)
    return pd.concat([temp,x],axis=1)
```
